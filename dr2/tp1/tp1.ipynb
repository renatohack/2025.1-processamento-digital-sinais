# Questão 1

Escreva um programa para:

### 1.1 – Gerar uma representação (um gráfico) de um sinal senoidal contínuo no tempo 
- Duração: 1 segundo
- Amplitude: 5
- Frequência: 3 Hz

### 1.2 – Gerar uma representação deste sinal amostrado no tempo.

### 1.3 – Qual a frequência de amostragem adotada?

### 1.4 – Porque foi adotado este valor?

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = 5
f = 3
phi = 0
```

```python
t_inicial = 0
t_final = 1
dt = 1e-3
```

```python
arr_T = np.arange(t_inicial, t_final + dt, dt)
```

```python
y = A * np.sin(2 * np.pi * f * arr_T + phi)
```

```python
plt.plot(arr_T, y)
plt.xlabel('t[s]')
plt.ylabel('y(t)')
plt.title('Sinal contínuo')
plt.grid()
plt.show()
```

```python
''' A frequência de amostragem foi selecionada como sendo 20x a frequência do sinal. Dado o Teorema de Nyquist que estipula
o limite mínimo da frequência como sendo f_a >= 2.f_s, escolhemos uma resolução 10x maior que o mínimo estipulado
para melhor clareza do sinal.
'''

f_amostragem = f * 20
```

```python
t_inicial = 0
t_final = 1
dt = 1 / f_amostragem
```

```python
arr_T_amostrado = np.arange(t_inicial, t_final + dt, dt)
```

```python
y_amostrado = A * np.sin(2 * np.pi * f * arr_T_amostrado + phi)
```

```python
plt.plot(arr_T_amostrado, y_amostrado, 'o')
plt.xlabel('t[s]')
plt.ylabel('y(t)')
plt.title('Sinal amostrado')
plt.grid()
plt.show()
```
