```python
# Nome: Soma_Senoides.py
# Função: Aproximar uma função por uma soma de senoides
# Data: 10/04/2023
# Autor: RD

import numpy as np
import matplotlib.pyplot as plt
import math

# 1 - Gerar sinal S(t)

pi= np.pi
h= 0.01
T= np.arange(-1,2*pi+1,h)
nT= len(T)
S= np.zeros(nT)
for i in range(0,nT):
    if (T[i]<0):
        S[i]= 0
    else:
        if(T[i]<pi):
            S[i]=1
        else:
            if(T[i]<2*pi):
                S[i]= -1
            else:
                S[i]= 0

T_pi= T/pi
plt.plot(T_pi,S)
plt.xlabel('t [pi]')
plt.ylabel('S(t)')
plt.grid()
plt.title('Gráfico do sinal S(t)')
plt.show()
```

```python
# 2- Geração das decomposições

Tn= np.arange(0,2*pi,h)
nTn= len(Tn)
Tn_pi= Tn/pi

Sn = np.zeros(nTn)
Sn_1 = np.zeros(nTn)
tamanho= 60
N= np.arange(1,tamanho,2)
nN= len(N)

for iN in range(0,nN):

    n= N[iN]
    print(n)

    for i in range(0,nTn):

        t= Tn[i]

        Sn[i]= Sn_1[i]+(4/pi)*(1/n)*(math.sin(n*t))

    plt.plot(T_pi,S,Tn_pi,Sn)
    plt.xlabel('t [pi]')
    plt.ylabel('Sinais')
    plt.grid()
    plt.legend(['S(t)','Sn(t)'])
    plt.title(f'Gráfico dos sinais para n = {iN+1}')
    plt.show()

    Sn_1 = Sn
```
