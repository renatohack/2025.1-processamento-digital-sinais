```python
# Nome: Exercicio2.py 
# Funo : Exerccio 2 da Seo Mtodos Estatsticos Aplicados em PDS 
# Data: 10/11/2022 
# Autor: RD 

import numpy as np
import matplotlib.pyplot as plt

nN = 1000
N = np.arange(0, nN)

# Sinais 
S1 = 1 * np.ones((nN, 1))
S2 = 2 * np.ones((nN, 1))
S3 = 3 * np.ones((nN, 1))

# Rudo 
mW = 0
sW = 2
W1 = np.random.normal(mW, sW, nN)
W2 = np.random.normal(mW, sW, nN)
W3 = np.random.normal(mW, sW, nN)

# Sinais com rudo aditivo 
Sw1 = np.zeros((nN, 1))
Sw2 = np.zeros((nN, 1))
Sw3 = np.zeros((nN, 1))

for i in range(nN):
    Sw1[i] = S1[i] + W1[i]
    Sw2[i] = S2[i] + W2[i]
    Sw3[i] = S3[i] + W3[i]

# Grficos 
plt.plot(N, Sw1)
plt.xlabel('n')
plt.ylabel('Sw1')
plt.title('S1 + W')
plt.grid()
plt.show()

plt.plot(N, Sw2)
plt.xlabel('n')
plt.ylabel('Sw2')
plt.title('S2 + W')
plt.grid()
plt.show()

plt.plot(N, Sw3)
plt.xlabel('n')
plt.ylabel('Sw3')
plt.title('S3 + W')
plt.grid()
plt.show()

# Mdia 
soma1 = 0
soma2 = 0
soma3 = 0

for iS in range(nN):
    soma1 += Sw1[iS]
    soma2 += Sw2[iS]
    soma3 += Sw3[iS]

mSw1 = soma1 / nN
mSw2 = soma2 / nN
mSw3 = soma3 / nN

# Varincia e desvio padro 
soma1 = 0
soma2 = 0
soma3 = 0

for iS in range(nN):
    soma1 += (Sw1[iS] - mSw1) ** 2
    soma2 += (Sw2[iS] - mSw2) ** 2
    soma3 += (Sw3[iS] - mSw3) ** 2

vSw1 = soma1 / (nN - 1)
vSw2 = soma2 / (nN - 1)
vSw3 = soma3 / (nN - 1)

sSw1 = np.sqrt(vSw1)
sSw2 = np.sqrt(vSw2)
sSw3 = np.sqrt(vSw3)

# Relao sinal rudo 
snr1 = mSw1 / sSw1
snr2 = mSw2 / sSw2
snr3 = mSw3 / sSw3

# Coeficiente da variao 
cv1 = (sSw1 / mSw1) * 100
cv2 = (sSw2 / mSw2) * 100
cv3 = (sSw3 / mSw3) * 100

# Resultados 
print('\nMdias:')
print(f'   mSw1 = {mSw1[0]:.4f}')
print(f'   mSw2 = {mSw2[0]:.4f}')
print(f'   mSw3 = {mSw3[0]:.4f}')

print('\nDesvios padro:')
print(f'   sSw1 = {sSw1[0]:.4f}')
print(f'   sSw2 = {sSw2[0]:.4f}')
print(f'   sSw3 = {sSw3[0]:.4f}')

print('\nRelao sinal-rudo (SNR):')
print(f'   snr1 = {snr1[0]:.4f}')
print(f'   snr2 = {snr2[0]:.4f}')
print(f'   snr3 = {snr3[0]:.4f}')

print('\nCoeficiente de variao:')
print(f'   cv1 = {cv1[0]:.2f} %')
print(f'   cv2 = {cv2[0]:.2f} %')
print(f'   cv3 = {cv3[0]:.2f} %')
```
