```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 100
t = np.arange(0, 1, 1/fs)
```

```python
sinal1 = np.sin(2 * np.pi * 1 * t)
sinal2 = np.sin(2 * np.pi * 5 * t)
sinal3 = np.sin(2 * np.pi * 10 * t)
```

```python
sinal_composto = sinal1 + sinal2 + sinal3
```

```python
def plot_sinal(t, sinal):
    plt.figure(figsize=(10, 4))
    plt.plot(t, sinal, label="Sinal Composto")
    plt.xlabel("Tempo (s)")
    plt.ylabel("Amplitude")
    plt.title("Sinal no Domnio do Tempo")
    plt.legend()
    plt.grid()

plot_sinal(t, sinal_composto)
```

```python
def calcular_espectro(sinal, fs):
    N = len(sinal)
    frequencias = np.fft.fftfreq(N, d=1/fs)[:N//2]
    espectro = np.abs(np.fft.fft(sinal))[:N//2] / N * 2
    return frequencias, espectro

def plot_espectro(frequencias, espectro):
    plt.figure(figsize=(10, 4))
    plt.stem(frequencias, espectro)
    plt.xlabel("Frequncia (Hz)")
    plt.ylabel("Magnitude")
    plt.title("Espectro de Frequncias")
    plt.grid()


frequencias, espectro = calcular_espectro(sinal_composto, fs)
plot_espectro(frequencias, espectro)
```

## **1. Qual dos sinais  o mais alto?**
O sinal de **10 Hz**  o mais alto.

## **2. Justifique sua resposta.**
A altura de um som est associada  sua frequncia: **frequncias mais altas produzem sons mais agudos** e **frequncias mais baixas produzem sons mais graves**. Como o sinal de 10 Hz possui a maior frequncia, ele  o mais alto entre os trs sinais.

## **3. Qual dos sinais apresenta maior intensidade?**
Todos os trs sinais possuem a **mesma intensidade**.

## **4. Explique o critrio utilizado para determinar a intensidade.**
A intensidade do sinal est relacionada  **energia transportada pelo sinal**, dada pela **amplitude** do sinal. Como todas as senoides possuem a mesma amplitude (1), a intensidade de cada uma individualmente  a mesma.

## **5. Defina o timbre de um dos sinais senoidais.**
O **timbre**  a caracterstica do som que permite distinguir dois sinais com a mesma altura (frequncia) e intensidade (amplitude). Um **sinal senoidal puro** (como os de 1 Hz, 5 Hz ou 10 Hz) tem um timbre **suave e puro**, pois contm apenas um nico componente de frequncia, sem harmnicos adicionais.

## **6. Escolha um dos sinais senoidais (1 Hz, 5 Hz ou 10 Hz) e descreva seu timbre.**
Se escolhermos, por exemplo, o **sinal de 5 Hz**, ele ter um timbre **puro e limpo**, sem componentes adicionais, soando como um tom nico sem variaes complexas.

## **7. Defina o timbre do sinal composto.**
O timbre do **sinal composto**  **mais complexo e rico**, pois ele contm **mltiplas frequncias (1 Hz, 5 Hz e 10 Hz)**.

## **8. Descreva como o timbre do sinal composto se diferencia dos sinais senoidais individuais.**
O timbre do **sinal composto** se diferencia porque contm **mltiplas frequncias simultaneamente**, criando um som **mais rico e mais complexo**, diferente dos demais senoidas, que so compostos por uma **nica frequncia**.
