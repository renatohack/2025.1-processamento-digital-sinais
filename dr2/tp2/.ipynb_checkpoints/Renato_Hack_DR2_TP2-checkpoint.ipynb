```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fs = 100
t = np.arange(0, 1, 1/fs)
```

```python
sinal1 = np.sin(2 * np.pi * 1 * t)
sinal2 = np.sin(2 * np.pi * 5 * t)
sinal3 = np.sin(2 * np.pi * 10 * t)
```

```python
sinal_composto = sinal1 + sinal2 + sinal3
```

```python
def plot_sinal(t, sinal):
    plt.figure(figsize=(10, 4))
    plt.plot(t, sinal, label="Sinal Composto")
    plt.xlabel("Tempo (s)")
    plt.ylabel("Amplitude")
    plt.title("Sinal no Domínio do Tempo")
    plt.legend()
    plt.grid()

plot_sinal(t, sinal_composto)
```

```python
def calcular_espectro(sinal, fs):
    N = len(sinal)
    frequencias = np.fft.fftfreq(N, d=1/fs)[:N//2]
    espectro = np.abs(np.fft.fft(sinal))[:N//2] / N * 2
    return frequencias, espectro

def plot_espectro(frequencias, espectro):
    plt.figure(figsize=(10, 4))
    plt.stem(frequencias, espectro)
    plt.xlabel("Frequência (Hz)")
    plt.ylabel("Magnitude")
    plt.title("Espectro de Frequências")
    plt.grid()


frequencias, espectro = calcular_espectro(sinal_composto, fs)
plot_espectro(frequencias, espectro)
```

# Análise dos Sinais Senoidais e do Sinal Composto

## **1. Qual dos sinais é o mais alto?**
O sinal de **10 Hz** é o mais alto.

## **2. Justifique sua resposta.**
A altura de um som está associada à sua frequência: **frequências mais altas produzem sons mais agudos** e **frequências mais baixas produzem sons mais graves**. Como o sinal de 10 Hz possui a maior frequência, ele é o mais alto entre os três sinais.

## **3. Qual dos sinais apresenta maior intensidade?**
Todos os três sinais possuem a **mesma intensidade**.

## **4. Explique o critério utilizado para determinar a intensidade.**
A intensidade do sinal está relacionada à **energia transportada pelo sinal**, dada pela **amplitude** do sinal. Como todas as senoides possuem a mesma amplitude (1), a intensidade de cada uma individualmente é a mesma.

## **5. Defina o timbre de um dos sinais senoidais.**
O **timbre** é a característica do som que permite distinguir dois sinais com a mesma altura (frequência) e intensidade (amplitude). Um **sinal senoidal puro** (como os de 1 Hz, 5 Hz ou 10 Hz) tem um timbre **suave e puro**, pois contém apenas um único componente de frequência, sem harmônicos adicionais.

## **6. Escolha um dos sinais senoidais (1 Hz, 5 Hz ou 10 Hz) e descreva seu timbre.**
Se escolhermos, por exemplo, o **sinal de 5 Hz**, ele terá um timbre **puro e limpo**, sem componentes adicionais, soando como um tom único sem variações complexas.

## **7. Defina o timbre do sinal composto.**
O timbre do **sinal composto** é **mais complexo e rico**, pois ele contém **múltiplas frequências (1 Hz, 5 Hz e 10 Hz)**.

## **8. Descreva como o timbre do sinal composto se diferencia dos sinais senoidais individuais.**
O timbre do **sinal composto** se diferencia porque contém **múltiplas frequências simultaneamente**, criando um som **mais rico e mais complexo**, diferente dos demais senoidas, que são compostos por uma **única frequência**.
